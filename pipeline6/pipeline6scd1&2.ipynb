```python
query='''
        create table  customers_infor (
        id int primary key,
        employeeid int,
        name varchar(100),
        email varchar(100),
        age int,
        city varchar(100)  
    )
    '''
```

```python
with engine.connect() as conn:
    conn.execute(text(query))
    conn.commit()
```
